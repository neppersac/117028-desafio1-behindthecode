```python
# Função para verificar estacionariedade
def verificar_estacionariedade(serie, nome_serie):
    # Teste de Dickey-Fuller aumentado (ADF test)
    resultado_adf = adfuller(serie)

    # Extraindo estatísticas do resultado do teste
    estatisticas_adf = resultado_adf[0]
    p_valor = resultado_adf[1]
    valores_criticos = resultado_adf[4]

    # Imprimindo os resultados
    print(f'Resultados do teste de Dickey-Fuller aumentado para a série {nome_serie}:')
    print(f'Estatística ADF: {estatisticas_adf}')
    print(f'p-valor: {p_valor}')
    print('Valores críticos:')
    for chave, valor in valores_criticos.items():
        print(f'   {chave}: {valor}')

    # Verificando a estacionariedade com base no p-valor
    if p_valor < 0.05:
        print(f'\nA série {nome_serie} é estacionária (rejeitamos a hipótese nula)')
    else:
        print(f'\nA série {nome_serie} não é estacionária (não rejeitamos a hipótese nula)')

```

```python
# Função para verificar estacionariedade e determinar o número de diferenciações necessárias
def verificar_estacionariedade(serie, nome_serie):
    # Teste de Dickey-Fuller aumentado (ADF test)
    resultado_adf = adfuller(serie)

    # Extraindo estatísticas do resultado do teste
    estatisticas_adf = resultado_adf[0]
    p_valor = resultado_adf[1]
    valores_criticos = resultado_adf[4]

    # Imprimindo os resultados
    print(f'Resultados do teste de Dickey-Fuller aumentado para a série {nome_serie}:')
    print(f'Estatística ADF: {estatisticas_adf}')
    print(f'p-valor: {p_valor}')
    print('Valores críticos:')
    for chave, valor in valores_criticos.items():
        print(f'   {chave}: {valor}')

    # Verificando a estacionariedade com base no p-valor
    if p_valor < 0.05:
        print(f'A série {nome_serie} é estacionária (rejeitamos a hipótese nula)')
        return 0  # Nenhuma diferença necessária
    else:
        # Calculando o número de diferenciações necessárias
        numero_diferencas = 0
        while p_valor >= 0.05:
            serie_diferenciada = serie.diff().dropna()  # Diferenciação da série
            resultado_adf = adfuller(serie_diferenciada)  # Teste ADF na série diferenciada
            p_valor = resultado_adf[1]
            numero_diferencas += 1

        print(f'A série {nome_serie} não é estacionária (não rejeitamos a hipótese nula)')
        print(f'Número de diferenciações necessárias: {numero_diferencas}')
        return numero_diferencas
```

```python
# Lista para armazenar o número de diferenciações necessárias para cada série
diferencas_necessarias = []

# Aplicando o teste de estacionariedade para cada série temporal
for coluna in df.columns[1:]:
    diferencas = verificar_estacionariedade(df[coluna], coluna)
    diferencas_necessarias.append(diferencas)

# Imprimindo o número de diferenciações necessárias para cada série
for i, coluna in enumerate(df.columns[1:]):
    print(f'Número de diferenciações necessárias para a série {coluna}: {diferencas_necessarias[i]}')
```
